```python
#solution to task no 1
class RealNumber:
    def __init__(self, r=0):
        self.__realValue = r

    def getRealValue(self):
        return self.__realValue

    def setRealValue(self, r):
        self.__realValue = r

    def __str__(self):
        return 'RealPart: ' + str(self.getRealValue())


class ComplexNumber(RealNumber):
    def __init__(self, num1=1, num2=1):
        super(ComplexNumber, self).__init__(float(num1))
        self.imaginary_value = float(num2)

    def __str__(self):
        return f'{super().__str__()}\nImaginaryPart: {self.imaginary_value}'


cn1 = ComplexNumber()
print(cn1)
print('---------')
cn2 = ComplexNumber(5, 7)
print(cn2)
```

```python
#solution to task no 2
# solution to task no 2
class RealNumber:
    def __init__(self, number=0):
        self.number = number

    def __add__(self, anotherRealNumber):
        return self.number + anotherRealNumber.number

    def __sub__(self, anotherRealNumber):
        return self.number - anotherRealNumber.number

    def __str__(self):
        return str(self.number)


class ComplexNumber(RealNumber):
    def __init__(self, num1, num2):
        super(ComplexNumber, self).__init__()
        self.num1 = num1
        self.num2 = num2

    def __str__(self):
        return f'{self.num1} + {self.num2}i'

    def __add__(self, other):
        return f'{self.num1 + other.num1.number} + {self.num2 + other.num2}i'

    def __sub__(self, other):
        return f'{self.num1 - other.num1.number} {self.num2 - other.num2}i'


r1 = RealNumber(3)
r2 = RealNumber(5)
print(r1 + r2)
cn1 = ComplexNumber(2, 1)
print(cn1)
cn2 = ComplexNumber(r1, 5)
print(cn2)
cn3 = cn1 + cn2
print(cn3)
cn4 = cn1 - cn2
print(cn4)
```

```python
#solution to task no 
class Account:
    def __init__(self, balance):
        self._balance = balance

    def getBalance(self):
        return self._balance


class CheckingAccount(Account):
    numberOfAccount = 0

    def __init__(self, balence=0.0):
        super(CheckingAccount, self).__init__(balence)
        CheckingAccount.numberOfAccount += 1

    def __str__(self):
        return f'Account Balance: {super().getBalance()}'


print('Number of Checking Accounts: ', CheckingAccount.numberOfAccount)
print(CheckingAccount())
print(CheckingAccount(100.00))
print(CheckingAccount(200.00))
print('Number of Checking Accounts: ', CheckingAccount.numberOfAccount)
```

```python
#solution to task no 5
class Exam:
    def __init__(self, marks):
        self.marks = marks
        self.time = 60

    def examSyllabus(self):
        return "Maths , English"

    def examParts(self):
        return "Part 1 - Maths\nPart 2 - English\n"


class ScienceExam(Exam):
    def __init__(self, marks, time, *subs):
        super(ScienceExam, self).__init__(marks)
        self.time = time
        self.subs = subs

    def examSyllabus(self):
        temp = ''
        for var in self.subs:
            temp += (' , ' + var)
        return super().examSyllabus() + temp

    def examParts(self):
        temp = ''
        for var in range(len(self.subs)):
            temp += f'Part {var + 3} - {self.subs[var]}\n'
        return super().examParts() + temp

    def __str__(self):
        return f'Marks: {self.marks} Time: {self.time} minutes Number of Parts: {len(self.subs) + 2}'


engineering = ScienceExam(100, 90, "Physics", "HigherMaths")
print(engineering)
print('----------------------------------')
print(engineering.examSyllabus())
print(engineering.examParts())
print('==================================')
architecture = ScienceExam(100, 120, "Physics", "HigherMaths", "Drawing")
print(architecture)
print('----------------------------------')
print(architecture.examSyllabus())
print(architecture.examParts())
```

```python
#solution to task no 6
class Shape3D:
    pi = 3.14159

    def __init__(self, name='Default', radius=0):
        self._area = 0
        self._name = name
        self._height = 'No need'
        self._radius = radius

    def calc_surface_area(self):
        return 2 * Shape3D.pi * self._radius

    def __str__(self):
        return "Radius: " + str(self._radius)


class Sphere(Shape3D):
    def __init__(self, name, radius):
        super(Sphere, self).__init__(name, radius)
        print('Shape name: Sphere, Area Formula: 4 * pi * r * r')

    def calc_surface_area(self):
        return super().calc_surface_area() * 2 * self._radius

    def __str__(self):
        return super().__str__() + f', Height: {self._height}\nArea: {self.calc_surface_area()}'


class Cylinder(Sphere, Shape3D):
    def __init__(self, name, radius, height):
        super(Sphere, self).__init__(name, radius)
        self._height = height
        print('Shape name: Cylinder, Area Formula: 2 * pi * r * (r + h)')

    def calc_surface_area(self):
        return super(Sphere, self).calc_surface_area() * (self._radius + self._height)


sph = Sphere('Sphere', 5)
print('----------------------------------')
sph.calc_surface_area()
print(sph)
print('==================================')
cyl = Cylinder('Cylinder', 5, 10)
print('----------------------------------')
cyl.calc_surface_area()
print(cyl)
```

```python
#solution to task no 7
class PokemonBasic:
    def __init__(self, name='Default', hp=0, weakness='None', type='Unknown'):
        self.name = name
        self.hit_point = hp
        self.weakness = weakness
        self.type = type

    def get_type(self):
        return 'Main type: ' + self.type

    def get_move(self):
        return 'Basic move: ' + 'Quick Attack'

    def __str__(self):
        return "Name: " + self.name + ", HP: " + str(self.hit_point) + ", Weakness: " + self.weakness


class PokemonExtra(PokemonBasic):
    def __init__(self, name, hp, weakness, main_type, *parameters):
        super(PokemonExtra, self).__init__(name, hp, weakness, main_type)
        self.parameters = parameters

    def __str__(self):
        return f'Name: {self.name}, HP: {self.hit_point}, Weakness: {self.weakness}'

    def get_type(self):
        if len(self.parameters) == 0:
            return super(PokemonExtra, self).get_type()
        else:
            return super(PokemonExtra, self).get_type() + f', Secondary type: {self.parameters[0]}'

    def get_move(self):
        if len(self.parameters) in [0, 1]:
            return super(PokemonExtra, self).get_move()
        else:
            temp = 'Other move: '
            for var in range(len(self.parameters[1])):
                if var == len(self.parameters[1]) - 1:
                    temp += self.parameters[1][var]
                else:
                    temp += self.parameters[1][var] + ', '
            return super(PokemonExtra, self).get_move() + '\n' + f'{temp}'


print('\n------------Basic Info:--------------')
pk = PokemonBasic()
print(pk)
print(pk.get_type())
print(pk.get_move())
print('\n------------Pokemon 1 Info:-------------')
charmander = PokemonExtra('Charmander', 39, 'Water', 'Fire')
print(charmander)
print(charmander.get_type())
print(charmander.get_move())
print('\n------------Pokemon 2 Info:-------------')
charizard = PokemonExtra('Charizard', 78, 'Water', 'Fire', 'Flying', ('Fire Spin', 'Fire Blaze'))
print(charizard)
print(charizard.get_type())
print(charizard.get_move())
```

```python
#solution to task no 9
class Pokemon:
    def __init__(self, p):
        self.pokemon = p
        self.pokemon_type = "Needs to be set"
        self.pokemon_weakness = "Needs to be set"

    def kind(self):
        return self.pokemon_type

    def weakness(self):
        return self.pokemon_weakness

    def what_am_i(self):
        print("I am a Pokemon.")


class Pikachu(Pokemon):
    def __init__(self, pokemon='Pikachu'):
        super(Pikachu, self).__init__(pokemon)
        self.pokemon_type = 'Electric'
        self.pokemon_weakness = 'Ground'

    def what_am_i(self):
        super(Pikachu, self).what_am_i()
        print('I am Pikachu.')


class Charmander(Pokemon):
    def __init__(self, pokemon='Charmander'):
        super(Charmander, self).__init__(pokemon)
        self.pokemon_type = 'Fire'
        self.pokemon_weakness = 'Water, Ground and Rock'

    def what_am_i(self):
        super(Charmander, self).what_am_i()
        print('I am Charmander.')


pk1 = Pikachu()
print("Pokemon:", pk1.pokemon)
print("Type:", pk1.kind())
print("Weakness:", pk1.weakness())
pk1.what_am_i()
print("========================")
c1 = Charmander()
print("Pokemon:", c1.pokemon)
print("Type:", c1.kind())
print("Weakness:", c1.weakness())
c1.what_am_i()
```

```python
#solution to task no 10
class Department:
    def __init__(self, s):
        self.semester = s
        self.name = "Default"
        self.id = -1

    def student_info(self):
        print("Name:", self.name)
        print("ID:", self.id)

    def courses(self, c1, c2, c3):
        print("No courses Approved yet!")


class CSE(Department):
    def __init__(self, name, stid, semister_year, department='CSE'):
        super(CSE, self).__init__(semister_year)
        self.name = name
        self.id = stid
        self.department = department

    def courses(self, c1, c2, c3):
        print(f'Courses Approved to this {self.department} student in {self.semester} semester :')
        print(c1)
        print(c2)
        print(c3)


class EEE(CSE):
    def __init__(self, name, stid, semister_year, department='EEE'):
        super(EEE, self).__init__(name, stid, semister_year, department)


s1 = CSE("Rahim", 16101328, "Spring2016")
s1.student_info()
s1.courses("CSE110", "MAT110", "ENG101")
print("==================")
s2 = EEE("Tanzim", 18101326, "Spring2018")
s2.student_info()
s2.courses("Mat110", "PHY111", "ENG101")
print("==================")
s3 = CSE("Rudana", 18101326, "Fall2017")
s3.student_info()
s3.courses("CSE111", "PHY101", "MAT120")
print("==================")
s4 = EEE("Zainab", 19201623, "Summer2019")
s4.student_info()
s4.courses("EEE201", "PHY112", "MAT120")
```
