```python
# solution of qustion no 1 
output_dict = {}
input_count = 0
while True:
    inputx = input('give your input : ')
    if inputx == 'STOP':
        break
    input_count += 1
    var = inputx.split(' ')
    if var[0] not in output_dict.keys():
        output_dict[var[0]] = [f'{var[0][0:2]}_{var[1]}_{input_count}']
    else:
        output_dict[var[0]].append(f'{var[0][0:2]}_{var[1]}_{input_count}')

print(output_dict)
```

```python
# solution of qustion no 2
class Chillox:
    menu = {}
    order_id = 66

    @classmethod
    def addBurger(cls, *burgs):  # each burger have 2 parts
        for var in range(len(burgs)):
            if var % 2 == 0:
                cls.menu[burgs[var]] = burgs[var + 1]

    @classmethod
    def placeOrderof(cls, customer):  # peramiter is a class
        cls.order_id += 1
        if ',' in customer.order:
            string = f'Order ID:{cls.order_id}\nCustomer name:{customer.name}\nOrder:'
            var = customer.order.split(',')
            for i in var:
                var1 = i[0:1]
                var2 = i[2:]
                string += f'\nBurger:{var2} , Quantity:{var1}'
                customer.bill += int(var1) * int(cls.menu[var2])
            print(string)
        else:
            var1 = customer.order[0:1]
            var2 = customer.order[2:]
            customer.bill += int(var1) * int(cls.menu[var2])
            print(f'Order ID:{cls.order_id}\nCustomer name:{customer.name}\nOrder:\nBurger:{var2} , Quantity:{var1}')


class Customer:
    order = None

    def __init__(self, name):
        self.name = name
        self.bill = 0

    @classmethod
    def setOrder(cls, order):  # order is a string . need separete parsts from it .
        cls.order = order

    def __str__(self):
        return f'Bill of Samin is {self.bill} taka'


print(f"It's been a good day!. \nToday, we have already served {Chillox.order_id} customers.")
print("**************************************************")
outlet1 = Chillox()
print("1.==========================================")
print("Burger Menu:")
print(Chillox.menu)
print("2.==========================================")
outlet1.addBurger("Chicken Burger", 180, "Chicken with Cheese", 200)
print("3.==========================================")
print("Burger Menu:")
print(Chillox.menu)
print("4.==========================================")
c1 = Customer("Samin")
c1.order = "2xChicken Burger"
outlet1.placeOrderof(c1)
print("5.==========================================")
outlet1.addBurger('Chicken Smoky BBQ Cheese', 225)
c2 = Customer("Rafi")
c2.setOrder("2xChicken with Cheese,3xChicken Smoky BBQ Cheese")
print("6.==========================================")
outlet1.placeOrderof(c2)
print("7.==========================================")
print(c1)
print("8.==========================================")
print(c2)
```

```python
# solution of qustion no 3
class Person:
    def __init__(self, name, contact):
        self.name = name
        self.contact = contact

    def __str__(self):
        s = f"Name: {self.name}\nContact: {self.contact}"
        return s


# Write your code here
class Doctor(Person):
    def __init__(self, name, contact, department, code):
        super(Doctor, self).__init__(name, contact)
        self.Speciality = department
        self.hospital = code

    def add_patients(self, *patient):  # tuple
        temp = {}
        self.patients = temp
        for var in patient:
            tempx = var.split(' ')
            x = f'{tempx[1]} {tempx[2]}'
            if tempx[0] not in temp.keys():
                temp[tempx[0]] = [x]
            else:
                temp[tempx[0]].append(x)



    def show_patient_list(self, place):
        if place not in self.patients.keys():
            print(f'{self.name} has not patients from {place}')
        else:
            print(f'Patients from{place}')
            copy = self.patients.copy()
            count= 0
            for var in copy.keys():
                xx = str(copy[var])[2:].split(',')
                if var == place:
                    print(f'Dhaka {copy[var][count][0:-1]}')
                    count+=1

    def __str__(self):
        return 'Doctor Details:\n'+super(Doctor, self).__str__()+f'\nSpeciality: {self.Speciality}\nHospital: {self.hospital}\nPatients List:\n{self.patients}'

class Engineer(Person):
    def __init__(self, name, contact, department):
        super(Engineer, self).__init__(name, contact)
        self.department = department

    def add_projects(self, *projects):
        temp = {}
        for var in projects:
            tempx = var.split(' ')
            x = f'{tempx[1]} {tempx[2]}'
            if tempx[0] not in temp.keys():
                temp[tempx[0]] = [x]
            else:
                temp[tempx[0]].append(x)
        self.project = temp
        
    def __str__(self):
        return 'Engineer Details:\n'+super(Engineer, self).__str__()+f'\nField: {self.department}\n{self.project}'

d = Doctor("Dr Strange", "4585632", "Neurosurgery", "DMC")
d.add_patients("Dhaka Abdur Rahim", "Dhaka John Cena", "Narayanganj John Wick", "Comilla CM Punk")
print("1.=========================================================")
print(d)
print("2.=========================================================")
d.show_patient_list("Dhaka")
print("3.=========================================================")
d.show_patient_list("Rajshahi")
print("4.=========================================================")
d.add_patients("Narayanganj Naruto Uzumaki")
print("5.=========================================================")
print(d)
print("6.=========================================================")
eng = Engineer("Howard Wolowitz", "982685", "Mechanical")
eng.add_projects("Mongol Tori Bangladesh", "Duburi Bangladesh", "Curiosity Rover USA", "Perseverence USA")
print("7.=========================================================")
print(eng)
```
