```python
class Anime:
    def __init__(self, name, rating, genre):
        self.name = name
        self.rating = rating
        self.genre = genre

    def add_character(self, *info):
        pass

    def __str__(self):
        s = f"Name: {self.name}\nRating: {self.rating}\nGenre: {self.genre}"
        return s


class Naruto(Anime):
    def __init__(self, name, rating, genre, release_date):
        super(Naruto, self).__init__(name, rating, genre)
        self.release_date = release_date

    def release_year(self):
        return f'{self.name} has been released in {self.release_date}!!!'

    def add_character(self, *info):
        temp = info
        temp_dict = {}
        for var in range(len(temp)):
            if var % 2 != 0:
                temp_dict[temp[var]] = []
        for var in range(len(temp)):
            if temp[var] in temp_dict.keys():
                temp_dict[temp[var]].append(temp[var - 1])
        self.characters = temp_dict

    def __str__(self):
        varx = f'Anime Details:\n' + super(Naruto, self).__str__() + f'\nRelease Year: {self.release_date}\nCharacters:'
        for var in self.characters.keys():
            varx += f'\n{var}: {self.characters[var]}'
        return varx


class MyHeroAcademia(Naruto):
    def __init__(self, name, rating, genare, season):
        relise_date = None
        super(MyHeroAcademia, self).__init__(name, rating, genare, relise_date)
        self.season = season

    def season_status(self):
        return f'{self.name} has premiered {self.season} seasons!!!'

    def add_character(self, *info):
        temp = info
        temp_dict = {}
        for var in range(len(temp)):
            if var % 2 == 0:
                temp_dict[temp[var]] = []
        for var in range(len(temp)):
            if temp[var] in temp_dict.keys():
                temp_dict[temp[var]].append(temp[var - 1])
        self.characters = temp_dict

    def __str__(self):
        varx = f'Anime Details:\n' + super(Naruto, self).__str__() + f'\nSeason Premiered: {self.season}\nCharacters:'
        for var in self.characters.keys():
            varx += f'\n{var}: {self.characters[var]}'
        return varx


# Write your codes here.
naruto = Naruto("Naruto", 10, "Adventure Fiction", 2007)
naruto.add_character("Naruto Uzumaki", "Main", "Itachi Uchiha", "Main", "Madara Uchiha", "Anti Hero", "Pain",
                     "Supporting", "Shikamaru Nara", "Supporting")
print('1.------------------------------------')
print(naruto.release_year())
print('2.------------------------------------')
print(naruto)
print('3.====================================')
my_hero_academia = MyHeroAcademia("My Hero Academia", 8, "Superhero Fiction", 5)
my_hero_academia.add_character("Supporting", "Uraraka", "Anti Hero", "Nomu", "Supporting", "Mirio", "Main", "Midoriya",
                               "Main", "Todoroki")
print('4.------------------------------------')
print(my_hero_academia.season_status())
print('5.------------------------------------')
print(my_hero_academia)
```
